## Data Visualization 
Viewership over Time

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Season'], df['Viewers (Millions)'], marker='o', linestyle='-')
plt.xlabel('Season')
plt.ylabel('Viewers (Millions)')
plt.title('Game of Thrones - Viewership over Seasons')
plt.grid(True)
plt.show()
```

Episode-to-episode Viewership Change

```python
df['Viewership Change'] = df['Viewers (Millions)'].diff()
plt.figure(figsize=(10, 6))
plt.plot(df['Episode'], df['Viewership Change'], marker='o', linestyle='-')
plt.xlabel('Episode')
plt.ylabel('Viewership Change (Millions)')
plt.title('Game of Thrones - Viewership Change per Episode')
plt.grid(True)
plt.show()
```
